```python
import pandas as pd

# Load train and test datasets
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# Check shapes
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# Preview first 5 rows
train_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Missing values in train
missing_train = train_df.isnull().sum().sort_values(ascending=False)
print("--- Missing values in Train ---")
print(missing_train[missing_train > 0])

# Missing values in test
missing_test = test_df.isnull().sum().sort_values(ascending=False)
print("\n--- Missing values in Test ---")
print(missing_test[missing_test > 0])

# Visualize missing values in train
plt.figure(figsize=(12,6))
sns.heatmap(train_df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap - Train Data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numeric columns
numeric_features = train_df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix
corr = numeric_features.corr()

# Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap - Numeric Features Only")
plt.show()

# Top correlations with SalePrice
top_corr = corr['SalePrice'].sort_values(ascending=False)[1:11]
print("Top 10 numeric features correlated with SalePrice:\n", top_corr)
```

```python
# Separate numeric and categorical columns in train
numeric_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = train_df.select_dtypes(include=['object']).columns

# Fill numeric columns with median
for col in numeric_cols:
    # Fill train set
    train_df[col] = train_df[col].fillna(train_df[col].median())
    
    # Fill test set only if column exists
    if col in test_df.columns:
        test_df[col] = test_df[col].fillna(test_df[col].median())

# Fill categorical columns with mode
for col in categorical_cols:
    # Fill train set
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
    
    # Fill test set only if column exists
    if col in test_df.columns:
        test_df[col] = test_df[col].fillna(test_df[col].mode()[0])

# Check remaining missing values
print("Missing values in Train after imputation:", train_df.isnull().sum().sum())
print("Missing values in Test after imputation:", test_df.isnull().sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

# Option 1: Label Encoding (for simplicity)
for col in categorical_cols:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
X_train = train_df.drop('SalePrice', axis=1)
y_train = train_df['SalePrice']

X_test = test_df.copy()  # Usually test set has no target
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Dictionary to store results
results = {}

# ------------------ KNN REGRESSOR ------------------
knn_model = KNeighborsRegressor(n_neighbors=5)  # you can tune n_neighbors
knn_model.fit(X_tr_scaled, y_tr)
y_val_pred_knn = knn_model.predict(X_val_scaled)
results['KNN'] = {
    'RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred_knn)),
    'R2': r2_score(y_val, y_val_pred_knn)
}

# ------------------ DECISION TREE REGRESSOR ------------------
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_tr_scaled, y_tr)
y_val_pred_dt = dt_model.predict(X_val_scaled)
results['Decision Tree'] = {
    'RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred_dt)),
    'R2': r2_score(y_val, y_val_pred_dt)
}

# ------------------ RANDOM FOREST REGRESSOR ------------------
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_tr_scaled, y_tr)
y_val_pred_rf = rf_model.predict(X_val_scaled)
results['Random Forest'] = {
    'RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred_rf)),
    'R2': r2_score(y_val, y_val_pred_rf)
}

# ------------------ ADABOOST REGRESSOR ------------------
ab_model = AdaBoostRegressor(n_estimators=100, random_state=42)
ab_model.fit(X_tr_scaled, y_tr)
y_val_pred_ab = ab_model.predict(X_val_scaled)
results['AdaBoost'] = {
    'RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred_ab)),
    'R2': r2_score(y_val, y_val_pred_ab)
}

# ------------------ SVM REGRESSOR ------------------
svm_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)  # you can tune C and epsilon
svm_model.fit(X_tr_scaled, y_tr)
y_val_pred_svm = svm_model.predict(X_val_scaled)
results['SVM'] = {
    'RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred_svm)),
    'R2': r2_score(y_val, y_val_pred_svm)
}

# ------------------ Display Results ------------------
print("Validation Results for All Models:")
for model_name, metrics in results.items():
    print(f"{model_name}: RMSE = {metrics['RMSE']:.2f}, R² = {metrics['R2']:.4f}")
```
